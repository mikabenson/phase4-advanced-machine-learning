# Introduction To Neural Networks 

An Artificial Neural Network (ANN) is a type of machine learning model that is inspired by how the human brain works. In the brain, neurons communicate with each other to process information and make decisions. Similarly, in a neural network, artificial neurons are connected to each other, and they work together to learn from data, recognize patterns, and make predictions or decisions.

The key idea is that these networks "learn" from data in a way that is similar to how we learn from experience.

### Neurons in a Neural Network

At the heart of a neural network are the neurons. A neuron takes input, processes it, and produces an output

- **Input:**

    The input to a neuron is a set of values (for example, pixel values in an image or features in a dataset). These inputs are multiplied by weights.

- **Weights:**
    
    Weights control how important each input is. In other words, they determine the strength of the connection between neurons.
- **Bias:**

    A bias is added to the weighted sum of the inputs. This allows the model to shift the activation function to better fit the data.

- **Output:** 

    After applying the bias and weights, an activation function is used to determine the final output.


   $$ 
   z = w_1 x_1 + w_2 x_2 + b 
   $$

## Neural Network Structure:

A typical neural network consists of the following layers:

- **Input Layer:**

    This is where the neural network receives the data. For example, in an image classification task, each neuron in the input layer represents a pixel in the image.
    Example: If the input is a 28x28 pixel image, the input layer will have 784 neurons (28  28 = 784).

- **Hidden Layers:**

    These layers are where the "magic" happens. They are called hidden because they are not exposed to the input or output directly.
    Each neuron in these layers processes the inputs from the previous layer (which could be the input layer or another hidden layer) and applies weights, biases, and activation functions.
    There can be multiple hidden layers. The more hidden layers, the "deeper" the neural network is, leading to the term Deep Learning when the network has many layers.

- **Output Layer:**

    The output layer provides the final predictions. For a classification problem, the number of neurons in this layer corresponds to the number of possible classes.
    Example: In a digit classification task (e.g., recognizing digits 0-9), there will be 10 neurons in the output layer, each representing one of the digits.

## Activation Function (f):

After calculating the weighted sum of inputs and adding the bias, the neuron applies an activation function to determine the output.

Common activation functions include:

- **Sigmoid:** 

    Outputs values between 0 and 1 (useful for binary classification).

- **ReLU (Rectified Linear Unit):** 

    Outputs 0 if input is less than 0, else outputs the input value (common in deep learning).

- **Tanh:** 

    Similar to sigmoid, but outputs between -1 and 1.

- **Softmax:** 

    Often used in the output layer for multi-class classification; converts the output into probabilities.

## How Neural Networks Learn

Neural networks are trained through a process that involves learning from input data and adjusting the network's internal parameters (weights and biases) to minimize the error between predictions and actual outcomes. This is achieved through a combination of **feedforward**, **backpropagation**, and **gradient descent**.

- ### Feedforward Process

    The feedforward process is the first step in how a neural network processes data and makes a prediction. Its the initial stage where the input is passed through the network to generate an output.


- #### The Feedforward Process

1. **Input Layer**:
   - The network receives data through the input layer. Lets say we are working with a dataset where each input consists of two features: x1 and x2.

2. **Hidden Layer**:
   - Each neuron in the hidden layer receives input from all neurons in the previous layer (the input layer). The input to each neuron is a weighted sum of the inputs.

3. **Mathematical Calculation of Weighted Sum**:
   - For each neuron, we calculate the weighted sum of inputs and add a **bias** term. This looks like this:



   $$ 
   z = w_1 x_1 + w_2 x_2 + b 
   $$

   
- `w1` and `w2` are the weights assigned to inputs `x1` and `x2`, respectively.
- `b` is the bias term that helps adjust the output of the neuron.

4. **Activation Function**:
- After calculating the weighted sum, an **activation function** (like sigmoid or ReLU) is applied to determine the output of the neuron. This output will then be passed to the next layer.


   
   $$ 
   a = f(z) = \text{sigmoid}(z) = \frac{1}{1 + e^{-z}} 
   $$

   - The result \( a \) is the output of the neuron, which will either be passed to the next hidden layer or, in the case of the output layer, be the final output of the network.

5. **Output Layer**:
   - The final output layer will give us the predicted value. For example, in a binary classification task (e.g., deciding whether an email is spam or not), the output will be a value between 0 and 1.

- ### Backpropagation

   **Backpropagation** is the key algorithm used for training neural networks. It is how the network "learns" from the mistakes it makes by adjusting its weights based on the error in the output.

- #### Backpropagation Proccess

   1. **Step 1: Make a Prediction**:
      - The input data is passed through the network (feedforward) to generate a prediction.

    2. **Step 2: Calculate the Error (Loss)**:
        - Once the network makes a prediction, we compare it to the actual value (the ground truth). The difference between the predicted output and the true value is called the **error**.
        
        - The **loss function** is used to quantify this error. For a regression problem, we might use **Mean Squared Error (MSE)**, and for classification, we use **Cross-Entropy Loss**.

            For MSE:
            ```
            L = 1/2 * (y_pred - y_true)^2
            ```

            - Where `y_pred` is the predicted value and `y_true` is the actual value.

    3. **Step 3: Backpropagation (Gradient Calculation)**:
        - To minimize the error, we need to adjust the weights of the network. The backpropagation algorithm calculates how much each weight contributed to the error using **gradients**.
        
        - The **gradient** is the rate of change of the loss function with respect to each weight. Mathematically, its the derivative of the loss function with respect to each weight.

    4. **Step 4: Update the Weights**:
        - Once the gradients are calculated, the weights are updated in the opposite direction of the gradient (since we want to minimize the loss).

            The weight update rule using gradient descent is:

            ```
            w_new = w_old - eta * (dL/dw)
            ```

            - `eta` is the **learning rate**.
            - `(dL/dw)` is the gradient of the loss function with respect to the weight `w`.

### Intuitive Flow:

1. **Input**  2. **Prediction**  3. **Error**  4. **Backpropagation (Gradient Calculation)**  5. **Update Weights**  6. **New Prediction**

- #### Gradient Descent

   **Gradient Descent** is the optimization technique used to minimize the loss function and improve the neural networks predictions. Its how the model "moves" toward better accuracy by adjusting its weights.

   ### Intuitive Understanding of Gradient Descent

   1. **What is Gradient Descent?**
      - Gradient descent is an iterative optimization algorithm used to minimize the loss function. By adjusting the weights step-by-step in the direction that reduces the loss, we gradually make the model better at predicting.

   2. **How Does Gradient Descent Work?**
      - The algorithm calculates the gradient of the loss function (the slope of the loss curve) and updates the weights in the opposite direction of the gradient. This is akin to moving downhill toward the lowest point of a loss curve (the global minimum).
      
      - Imagine you are standing on a hill. The gradient tells you the steepest direction to go. You move a little in that direction, and the process repeats, getting closer to the "bottom" of the hill (the minimum loss).

   3. **Learning Rate**:
      - The **learning rate** (`eta`) determines how big a step the algorithm takes in the direction of the gradient. 
      - If the learning rate is too large, you might overshoot the minimum.
      - If the learning rate is too small, the algorithm will converge too slowly and take a long time to reach the minimum.

      - A **good learning rate** ensures that the network moves efficiently toward the minimum loss.

   ### Mathematical Representation of Gradient Descent:

      The update rule is:



    w_new = w_old - eta * grad_L(w)


- `eta` represents the **learning rate**.
- `grad_L(w)` represents the gradient of the loss function with respect to the weights.

The algorithm repeats this process over many iterations (or **epochs**) until the loss reaches a minimum.
